## <span style="color:#6b99b0"><i><u> Data Cleaning </u></i></span>

```python
# There is no duplicate or missing values in the dataset of offline
dataframe.fillna(dataframe.mean(numeric_only=True), inplace=True)
```

```python
dataframe.drop_duplicates(inplace=True)
```

```python
dataframe.dropna(subset=['Attrition'], inplace=True)
```

## <span style="color:grey"> <u> Creation of Input and Output Features </u></span>

```python
# create input and output data
Features = dataframe.drop('Attrition', axis=1)
Labels = dataframe['Attrition']
Features
```

```python
Labels
```

## Conversion of features to numerical values

```python
categorical_columns = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']

# Convert each column to categorical type
for col in categorical_columns:
    Features[col] = Features[col].astype('category')
```

```python
# One-hot encode the data using pandas get_dummies
Features = pd.get_dummies(Features)
```
